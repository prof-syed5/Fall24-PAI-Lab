```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
def fillnaFloatMean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())

columns = ['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
fillnaFloatMean(columns)
```

```python
def fillnaObjectMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['PassengerId','HomePlanet','CryoSleep','Cabin','Destination','VIP','Name']
fillnaObjectMode(columns)
```

```python

def changeFloatToInt(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
changeFloatToInt(columns)
```

```python
def convertObjToInt(cols):
    for i in cols:
        data = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))    
        df[i] = data_LE.transform(df[i])
        
columns = ['PassengerId','HomePlanet','CryoSleep','Cabin','Destination','VIP','Name']
convertObjToInt(columns)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
def drop_columns(df: pd.DataFrame, columns: list):
    return df.drop(columns=[col for col in columns if col in df.columns], inplace=False)
cols_to_drop = ['PassengerId','Name','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
df = drop_columns(df, cols_to_drop)
```

```python
df.info()
```

```python
df.to_csv('processed1.csv', index=False, header=True)
```
