# **Water jug problem**

```python
def display_menu():
    print(" WATER JUG MENU ")
    print("1. Fill Jug X")
    print("2. Fill Jug Y")
    print("3. Empty Jug X")
    print("4. Empty Jug Y")
    print("5. Pour X -> Y (until X empty)")
    print("6. Pour X -> Y (until Y full)")
    print("7. Pour Y -> X (until Y empty)")
    print("8. Pour Y -> X (until X full)")
    print("9. Exit")


def water_jug_game():
    x = 0
    y = 0
    capacity_x = 4
    capacity_y = 3
    target = 2

    while True:
        print(f"\nCurrent State => Jug X: {x}L, Jug Y: {y}L")
        display_menu()

        try:
            choice = int(input("Select an option: "))
        except ValueError:
            print("Please enter a valid number.")
            continue

        if choice == 1:
            x = capacity_x

        elif choice == 2:
            y = capacity_y

        elif choice == 3:
            x = 0

        elif choice == 4:
            y = 0

        elif choice == 5:
            transfer = min(x, capacity_y - y)
            x -= transfer
            y += transfer

        elif choice == 6:
            transfer = min(x, capacity_y - y)
            x -= transfer
            y += transfer

        elif choice == 7:
            transfer = min(y, capacity_x - x)
            y -= transfer
            x += transfer

        elif choice == 8:
            transfer = min(y, capacity_x - x)
            y -= transfer
            x += transfer

        elif choice == 9:
            print("Game terminated.")
            break

        else:
            print("Invalid option! Try again.")
            continue

        if x == target or y == target:
            print(f"\nFinal State => Jug X: {x}L, Jug Y: {y}L")
            print("Goal achieved")
            break

water_jug_game()
```
